```python
merged_stat = pd.concat([away1, home1], ignore_index=True)

baseball_cols2 = ['名稱','出局數','面對打者','安打','全壘打','保送', '故意四壞','觸身', '三振', '暴投','投手犯規','失分','責失'] 
# 輸出合併後的資料表
merged_stat = merged_stat[baseball_cols2]
merged_stat
```

```python
agg_dict = {
    "出局數": "sum",
    "面對打者": "sum",
    "安打": "sum",
    "全壘打": "sum",
    "保送": "sum",
    "故意四壞": "sum",
    "觸身": "sum",
    "三振": "sum",
    "暴投": "sum",
    "投手犯規": "sum",
    "失分": "sum",
    "責失": "sum"
}
pitcher_total = merged_stat.groupby("名稱").agg(agg_dict).reset_index()
pitcher_total
```

```python
pitcher_total["IP"] = pitcher_total["出局數"] / 3

# 計算總 IP
#total_ip = pitcher_total["IP"].sum()
#pitcher_total["總局數"] = pitcher_total["出局數"].sum()

# 計算 ERA（失分 / 總 IP * 9）
pitcher_total["ERA"] = (pitcher_total["責失"] / pitcher_total["IP"]) * 9

# 計算 FIP
pitcher_total["FIP"] = 0.33 * (3 *pitcher_total["全壘打"] + 3 * pitcher_total["保送"] + 2 * pitcher_total["觸身"] + 3 * pitcher_total["暴投"] - 2 * pitcher_total["三振"]) / pitcher_total["IP"] + 3.2

# 計算 K%（三振 / 面對打者）
pitcher_total["K%"] = (pitcher_total["三振"] / pitcher_total["面對打者"]) * 100

# 計算 BB%（保送 / 面對打者）
pitcher_total["BB%"] = (pitcher_total["保送"] / pitcher_total["面對打者"]) * 100

# 移除指定的欄位
pitcher_total.drop(columns=["出局數","面對打者", "安打", "全壘打", "保送", "故意四壞", "觸身", "三振", "暴投", "投手犯規", "失分", "責失"], inplace=True)

# 列出計算結果
pitcher_total
```

```python
# 先將兩個資料表合併，根據共同的"投手"（"名稱"）欄位
merged_pitcher = pd.merge(pitcher_summary, pitcher_total, left_on="投手", right_on="名稱", how="inner")

# 移除重複的"名稱"欄位
merged_pitcher.drop(columns=["所屬球隊"], inplace=True)

# 列出合併後的結果
merged_pitcher
```
