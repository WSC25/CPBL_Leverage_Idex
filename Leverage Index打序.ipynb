```python
excel_file_path = 'C:/Users/wei/Documents/實習/RebasCPBL/中職2023年-逐球數據-20230825.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name3 = 'AWAY-打席資訊'  # 請替換成你要載入的工作表名稱
sheet_name4 = 'HOME-打席資訊'
```

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name3)
away
```

```python
away_play = away[baseball_cols]
away_play
```

```python
away_play = pd.DataFrame(away_play)

# 计算客場分差并添加到数据框中
away_play["分差"] = away_play["客場分"] - away_play["主場分"]

# 将负数显示为负数
away_play["分差"] = away_play["分差"].apply(lambda x: x if x >= 0 else f'({-x})')
```

```python
away_play
```

```python
li_away_df = pd.DataFrame(li_away)
away_play_df = pd.DataFrame(away_play)
```

```python
li_away_df
```

```python
# Specify the id_vars (columns to keep as identifiers)
id_vars = ["局數", "壘包狀態", "出局數"]

# Reshape the data using melt
li_away_df = pd.melt(li_away_df, id_vars=id_vars, var_name="分差", value_name="li值")

# Filter out rows with "li值" equal to 0
li_away_df = li_away_df[li_away_df["li值"] != 0]

# Convert "分差" to integer
li_away_df["分差"] = li_away_df["分差"].astype(int)

# Reset index
li_away_df.reset_index(drop=True, inplace=True)

# Output the result
#li_away_df
```

```python
li_away_df['分差'] = li_away_df['分差'].apply(lambda x: f'({abs(x)})' if x < 0 else x)
li_away_df
```

```python
li_values = []

# 遍历 away_play 数据表中的每一行
for index, row in away_play_df.iterrows():
    inning = row["局數"]
    bases = row["壘包狀態"]
    outs = row["出局數"]
    score = row['分差']
    
    # 在 li_away 表中找到匹配的行
    li_row = li_away_df[(li_away_df["局數"] == inning) & (li_away_df["壘包狀態"] == bases) & (li_away_df["出局數"] == outs) & (li_away_df["分差"] == score)]
    
    if not li_row.empty:
        li_value = li_row.iloc[0]["li值"]
        li_values.append(li_value)
    else:
        li_values.append(None)

# 将 li_values 列添加到 away_play 数据表中
away_play_df["li值"] = li_values

away_play_df
```
